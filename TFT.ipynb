```python
# ==========================================
# 1. Load Libraries
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Set random seeds
torch.manual_seed(42)
np.random.seed(42)

# ==========================================
# 2. Load Dataset from CSV
# ==========================================
df = pd.read_csv('SeoulBikeData.csv', encoding='latin1')
print("CSV Loaded Successfully!")
target_col = 'Rented Bike Count'

# ==========================================
# 3. Feature Engineering Function
# ==========================================
def add_engineered_features(df):
    # --- Fix Date Column ---
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    
    # Extract date parts
    df['Day'] = df['Date'].dt.day
    df['Month'] = df['Date'].dt.month
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
    
    # --- Cyclical Encoding ---
    df['Hour_Sin'] = np.sin(2 * np.pi * df['Hour'] / 24)
    df['Hour_Cos'] = np.cos(2 * np.pi * df['Hour'] / 24)
    df['DayOfWeek_Sin'] = np.sin(2 * np.pi * df['DayOfWeek'] / 7)
    df['DayOfWeek_Cos'] = np.cos(2 * np.pi * df['DayOfWeek'] / 7)
    df['Month_Sin'] = np.sin(2 * np.pi * df['Month'] / 12)
    df['Month_Cos'] = np.cos(2 * np.pi * df['Month'] / 12)
    
    # --- Weather Interaction Features ---
    df['feels_like_temp'] = df['Temperature(°C)'] - (0.1 * df['Humidity(%)'])
    
    # --- Flags (Weekend, Peak Hours) ---
    df['is_weekend'] = df['DayOfWeek'].isin([5, 6]).astype(int)
    df['is_morning_peak'] = df['Hour'].isin([7, 8, 9, 10]).astype(int)
    df['is_evening_peak'] = df['Hour'].isin([17, 18, 19, 20]).astype(int)
    df['is_low_demand_hour'] = df['Hour'].between(0, 5).astype(int)
    
    # --- Binary Conversion ---
    df['Holiday'] = df['Holiday'].map({'No Holiday': 0, 'Holiday': 1})
    df['Functioning Day'] = df['Functioning Day'].map({'Yes': 1, 'No': 0})
    
    # --- Weather binaries ---
    df['Has_Rain'] = (df['Rainfall(mm)'] > 0).astype(int)
    df['Has_Snow'] = (df['Snowfall (cm)'] > 0).astype(int)
    df['Good_Visibility'] = (df['Visibility (10m)'] > 1000).astype(int)
    df['Is_Sunny'] = (df['Solar Radiation (MJ/m2)'] > 0).astype(int)
    
    # --- Target transformation ---
    df['Rented_Bike_Count_Log'] = np.log1p(df['Rented Bike Count'])
    
    # Drop NaNs
    df = df.dropna().reset_index(drop=True)
    
    return df

# ==========================================
# 4. Apply Feature Engineering
# ==========================================
df = add_engineered_features(df)
print("New dataframe shape:", df.shape)
print(df.head())

# ==========================================
# 5. EDA Section
# ==========================================
print("\n" + "="*70)
print("EXPLORATORY DATA ANALYSIS")
print("="*70)

# ---- Overview ----
print("\nShape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nMissing values:\n", df.isnull().sum())

# ---- Summary ----
print("\nSummary Statistics:")
print(df.describe())

# ---- Outlier Detection ----
def outlier_count(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    return len(outliers)

print("\nOutlier Detection:")
for col in ['Rented Bike Count', 'Wind speed (m/s)', 'Solar Radiation (MJ/m2)', 
            'Rainfall(mm)', 'Snowfall (cm)']:
    print(f"{col} outliers:", outlier_count(col))

# ---- Distribution ----
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Rented Bike Count'], bins=30, kde=True)
plt.title('Bike Count Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Temperature(°C)'])
plt.title('Temperature Boxplot')
plt.tight_layout()
plt.show()

# ---- Correlation Matrix ----
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()

print("\nTop Correlations with Rented Bike Count:")
print(corr['Rented Bike Count'].sort_values(ascending=False).head(15))

# ---- Seasonality Plots ----
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.lineplot(
    data=df.groupby('Hour')['Rented Bike Count'].mean().reset_index(),
    x='Hour', y='Rented Bike Count'
)
plt.title('Average Demand by Hour')

plt.subplot(1, 3, 2)
sns.boxplot(x='Seasons', y='Rented Bike Count', data=df)
plt.title('Demand by Season')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
sns.lineplot(
    data=df.groupby('DayOfWeek')['Rented Bike Count'].mean().reset_index(),
    x='DayOfWeek', y='Rented Bike Count', marker='o'
)
plt.title('Average Demand by Day of Week')
plt.xlabel('Day (0=Monday, 6=Sunday)')
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(16, 12))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=False, fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

# ==========================================
# 6. DATA PREPARATION FOR TFT
# ==========================================
print("\n" + "="*70)
print("DATA PREPARATION FOR TFT MODEL")
print("="*70)

# Encode categorical variables
print("\nEncoding categorical variables...")
label_encoders = {}
for col in ['Seasons']:
    if col in df.columns:
        le = LabelEncoder()
        df[col + '_Encoded'] = le.fit_transform(df[col].astype(str))
        label_encoders[col] = le

# Define feature groups
static_features = ['Seasons_Encoded', 'Holiday', 'Functioning Day']

temporal_numeric_features = [
    'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)',
    'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)',
    'feels_like_temp'
]

temporal_cyclical_features = [
    'Hour_Sin', 'Hour_Cos', 'DayOfWeek_Sin', 'DayOfWeek_Cos',
    'Month_Sin', 'Month_Cos'
]

temporal_binary_features = [
    'is_weekend', 'is_morning_peak', 'is_evening_peak', 'is_low_demand_hour',
    'Has_Rain', 'Has_Snow', 'Good_Visibility', 'Is_Sunny'
]

temporal_features = temporal_numeric_features + temporal_cyclical_features + temporal_binary_features

print(f"Static features: {len(static_features)}")
print(f"Temporal features: {len(temporal_features)}")

# Scale numerical features
print("\nScaling features...")
scaler = StandardScaler()
df[temporal_numeric_features] = scaler.fit_transform(df[temporal_numeric_features])

# Sort by date and hour (CRITICAL for temporal split)
df = df.sort_values(['Date', 'Hour']).reset_index(drop=True)
print(f"\nData sorted by Date and Hour")
print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")

# ==========================================
# 7. CREATE TIME SERIES SEQUENCES
# ==========================================
print("\n" + "="*70)
print("CREATING TIME SERIES SEQUENCES")
print("="*70)

LOOKBACK_WINDOW = 72  # 3 days
FORECAST_HORIZON = 24  # 24 hours ahead

def create_sequences(data, lookback, horizon, static_cols, temporal_cols, target_col):
    X_static = []
    X_temporal = []
    y = []
    
    for i in range(lookback, len(data) - horizon + 1):
        # Static features
        static_seq = data[static_cols].iloc[i].values
        
        # Temporal features (past observations)
        temporal_seq = data[temporal_cols].iloc[i-lookback:i].values
        
        # Target (future values)
        target_seq = data[target_col].iloc[i:i+horizon].values
        
        X_static.append(static_seq)
        X_temporal.append(temporal_seq)
        y.append(target_seq)
    
    return np.array(X_static), np.array(X_temporal), np.array(y)

print(f"Creating sequences with lookback={LOOKBACK_WINDOW}, horizon={FORECAST_HORIZON}...")

X_static, X_temporal, y = create_sequences(
    df,
    LOOKBACK_WINDOW,
    FORECAST_HORIZON,
    static_features,
    temporal_features,
    'Rented_Bike_Count_Log'
)

print(f"X_static shape: {X_static.shape}")
print(f"X_temporal shape: {X_temporal.shape}")
print(f"y shape: {y.shape}")

# ==========================================
# 8. TEMPORAL TRAIN-TEST SPLIT (80/20)
# ==========================================
print("\n" + "="*70)
print("TRAIN-TEST SPLIT (TEMPORAL)")
print("="*70)

# Calculate split index for 80/20
split_idx = int(0.8 * len(X_static))

X_static_train = X_static[:split_idx]
X_static_test = X_static[split_idx:]

X_temporal_train = X_temporal[:split_idx]
X_temporal_test = X_temporal[split_idx:]

y_train = y[:split_idx]
y_test = y[split_idx:]

print(f"Train size: {len(X_static_train)} samples")
print(f"Test size: {len(X_static_test)} samples")
print(f"Train percentage: {(len(X_static_train)/len(X_static))*100:.1f}%")
print(f"Test percentage: {(len(X_static_test)/len(X_static))*100:.1f}%")

# Show date ranges for train/test
train_end_idx = split_idx + LOOKBACK_WINDOW
test_start_idx = train_end_idx
print(f"\nTrain data covers: {df['Date'].iloc[0]} to {df['Date'].iloc[train_end_idx]}")
print(f"Test data covers: {df['Date'].iloc[test_start_idx]} to {df['Date'].iloc[-1]}")

# ==========================================
# 9. PYTORCH DATASET AND DATALOADER
# ==========================================
class BikeDataset(Dataset):
    def __init__(self, X_static, X_temporal, y):
        self.X_static = torch.FloatTensor(X_static)
        self.X_temporal = torch.FloatTensor(X_temporal)
        self.y = torch.FloatTensor(y)
    
    def __len__(self):
        return len(self.X_static)
    
    def __getitem__(self, idx):
        return self.X_static[idx], self.X_temporal[idx], self.y[idx]

train_dataset = BikeDataset(X_static_train, X_temporal_train, y_train)
test_dataset = BikeDataset(X_static_test, X_temporal_test, y_test)

BATCH_SIZE = 64
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"\nBatch size: {BATCH_SIZE}")
print(f"Number of training batches: {len(train_loader)}")
print(f"Number of test batches: {len(test_loader)}")

# ==========================================
# 10. TFT MODEL ARCHITECTURE
# ==========================================
print("\n" + "="*70)
print("BUILDING ENHANCED TFT MODEL")
print("="*70)

class VariableSelectionNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout=0.1):
        super().__init__()
        self.hidden_dim = hidden_dim
        
        self.feature_transform = nn.Linear(input_dim, hidden_dim)
        
        self.variable_selection = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.Tanh(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, input_dim),
            nn.Softmax(dim=-1)
        )
        
        self.output_layer = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        transformed = self.feature_transform(x)
        weights = self.variable_selection(transformed)
        weighted_features = x * weights
        output = self.output_layer(self.feature_transform(weighted_features))
        return output, weights

class GatedResidualNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout=0.1):
        super().__init__()
        
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.elu = nn.ELU()
        self.fc2 = nn.Linear(hidden_dim, output_dim)
        self.dropout = nn.Dropout(dropout)
        
        self.gate = nn.Sequential(
            nn.Linear(output_dim, output_dim),
            nn.Sigmoid()
        )
        
        if input_dim != output_dim:
            self.skip = nn.Linear(input_dim, output_dim)
        else:
            self.skip = None
        
        self.layer_norm = nn.LayerNorm(output_dim)
    
    def forward(self, x):
        out = self.fc1(x)
        out = self.elu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        
        gate = self.gate(out)
        out = out * gate
        
        if self.skip is not None:
            x = self.skip(x)
        
        out = self.layer_norm(out + x)
        return out

class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads, dropout=0.1):
        super().__init__()
        assert d_model % num_heads == 0
        
        self.d_model = d_model
        self.num_heads = num_heads
        self.d_k = d_model // num_heads
        
        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.W_o = nn.Linear(d_model, d_model)
        
        self.dropout = nn.Dropout(dropout)
        self.layer_norm = nn.LayerNorm(d_model)
    
    def forward(self, query, key, value, mask=None):
        batch_size = query.size(0)
        
        Q = self.W_q(query).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        K = self.W_k(key).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        V = self.W_v(value).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        
        scores = torch.matmul(Q, K.transpose(-2, -1)) / np.sqrt(self.d_k)
        
        if mask is not None:
            scores = scores.masked_fill(mask == 0, -1e9)
        
        attention_weights = torch.softmax(scores, dim=-1)
        attention_weights = self.dropout(attention_weights)
        
        context = torch.matmul(attention_weights, V)
        context = context.transpose(1, 2).contiguous().view(batch_size, -1, self.d_model)
        output = self.W_o(context)
        output = self.layer_norm(output + query)
        
        return output, attention_weights

class EnhancedTemporalFusionTransformer(nn.Module):
    def __init__(self, static_dim, temporal_dim, hidden_dim=128, lstm_layers=2,
                 num_heads=8, dropout=0.1, forecast_horizon=24):
        super().__init__()
        
        self.static_dim = static_dim
        self.temporal_dim = temporal_dim
        self.hidden_dim = hidden_dim
        self.forecast_horizon = forecast_horizon
        
        # Variable Selection Networks
        self.static_vsn = VariableSelectionNetwork(
            static_dim, hidden_dim, hidden_dim, dropout
        )
        self.temporal_vsn = VariableSelectionNetwork(
            temporal_dim, hidden_dim, hidden_dim, dropout
        )
        
        # Static Covariate Encoder
        self.static_encoder = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, hidden_dim)
        )
        
        # LSTM Layers
        self.lstm1 = nn.LSTM(
            hidden_dim, hidden_dim, num_layers=lstm_layers,
            batch_first=True, dropout=dropout if lstm_layers > 1 else 0
        )
        
        self.lstm2 = nn.LSTM(
            hidden_dim, hidden_dim, num_layers=1, batch_first=True
        )
        
        # Multi-Head Self-Attention
        self.self_attention = MultiHeadAttention(hidden_dim, num_heads, dropout)
        
        # Gated Residual Networks
        self.grn1 = GatedResidualNetwork(hidden_dim, hidden_dim, hidden_dim, dropout)
        self.grn2 = GatedResidualNetwork(hidden_dim, hidden_dim, hidden_dim, dropout)
        
        # Decoder LSTM
        self.decoder_lstm = nn.LSTM(hidden_dim, hidden_dim, num_layers=1, batch_first=True)
        
        # Output layers
        self.output_layer = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim // 2, 1)
        )
        
        # Quantile outputs (for uncertainty)
        self.quantile_output = nn.ModuleList([
            nn.Linear(hidden_dim, 1) for _ in range(3)
        ])
    
    def forward(self, static_features, temporal_features):
        batch_size = static_features.size(0)
        seq_len = temporal_features.size(1)
        
        # Variable Selection
        static_encoded, static_weights = self.static_vsn(static_features)
        
        temporal_encoded_list = []
        for t in range(seq_len):
            temp_enc, _ = self.temporal_vsn(temporal_features[:, t, :])
            temporal_encoded_list.append(temp_enc.unsqueeze(1))
        temporal_encoded = torch.cat(temporal_encoded_list, dim=1)
        
        # Static encoding
        static_context = self.static_encoder(static_encoded)
        
        # LSTM processing
        lstm_out1, (h1, c1) = self.lstm1(temporal_encoded)
        lstm_out2, (h2, c2) = self.lstm2(lstm_out1)
        
        # Self-Attention
        attention_out, attention_weights = self.self_attention(
            lstm_out2, lstm_out2, lstm_out2
        )
        
        # Gated Residual Networks
        grn_out1 = self.grn1(attention_out)
        
        static_context_expanded = static_context.unsqueeze(1).repeat(1, seq_len, 1)
        combined = grn_out1 + static_context_expanded
        
        grn_out2 = self.grn2(combined)
        
        # Decoder
        context = grn_out2[:, -1, :].unsqueeze(1)
        decoder_input = context.repeat(1, self.forecast_horizon, 1)
        decoder_out, _ = self.decoder_lstm(decoder_input, (h2, c2))
        
        # Predictions
        predictions = self.output_layer(decoder_out).squeeze(-1)
        quantile_preds = [q_layer(decoder_out).squeeze(-1) for q_layer in self.quantile_output]
        
        return predictions, attention_weights, static_weights, quantile_preds

# Initialize model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

model = EnhancedTemporalFusionTransformer(
    static_dim=len(static_features),
    temporal_dim=len(temporal_features),
    hidden_dim=128,
    lstm_layers=2,
    num_heads=8,
    dropout=0.1,
    forecast_horizon=FORECAST_HORIZON
).to(device)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")

# ==========================================
# 11. TRAINING SETUP
# ==========================================
print("\n" + "="*70)
print("TRAINING CONFIGURATION")
print("="*70)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=5
)

def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    
    progress_bar = tqdm(dataloader, desc="Training")
    for batch_idx, (static, temporal, target) in enumerate(progress_bar):
        static = static.to(device)
        temporal = temporal.to(device)
        target = target.to(device)
        
        optimizer.zero_grad()
        predictions, _, _, _ = model(static, temporal)
        
        loss = criterion(predictions, target)
        loss.backward()
        
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        
        total_loss += loss.item()
        progress_bar.set_postfix({'loss': loss.item()})
    
    return total_loss / len(dataloader)

def validate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    all_predictions = []
    all_targets = []
    
    with torch.no_grad():
        for static, temporal, target in dataloader:
            static = static.to(device)
            temporal = temporal.to(device)
            target = target.to(device)
            
            predictions, _, _, _ = model(static, temporal)
            loss = criterion(predictions, target)
            
            total_loss += loss.item()
            all_predictions.append(predictions.cpu().numpy())
            all_targets.append(target.cpu().numpy())
    
    all_predictions = np.concatenate(all_predictions, axis=0)
    all_targets = np.concatenate(all_targets, axis=0)
    
    return total_loss / len(dataloader), all_predictions, all_targets

# ==========================================
# 12. TRAINING LOOP
# ==========================================
print("\n" + "="*70)
print("STARTING TRAINING")
print("="*70)

NUM_EPOCHS = 50
EARLY_STOPPING_PATIENCE = 10

train_losses = []
val_losses = []
best_val_loss = float('inf')
patience_counter = 0

for epoch in range(NUM_EPOCHS):
    print(f"\nEpoch {epoch+1}/{NUM_EPOCHS}")
    print("-" * 50)
    
    # Train
    train_loss = train_epoch(model, train_loader, criterion, optimizer, device)
    train_losses.append(train_loss)
    
    # Validate
    val_loss, val_preds, val_targets = validate(model, test_loader, criterion, device)
    val_losses.append(val_loss)
    
    scheduler.step(val_loss)
    
    # Calculate metrics (convert back from log scale)
    val_preds_original = np.expm1(val_preds)
    val_targets_original = np.expm1(val_targets)
    
    mae = np.mean(np.abs(val_preds_original - val_targets_original))
    rmse = np.sqrt(np.mean((val_preds_original - val_targets_original) ** 2))
    mape = np.mean(np.abs((val_targets_original - val_preds_original) / (val_targets_original + 1))) * 100
    
    print(f"Train Loss: {train_loss:.4f}")
    print(f"Val Loss: {val_loss:.4f}")
    print(f"MAE: {mae:.2f} | RMSE: {rmse:.2f} | MAPE: {mape:.2f}%")
    
    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_loss': val_loss,
            'mae': mae,
            'rmse': rmse
        }, 'best_tft_model.pth')
        print("✓ Best model saved!")
    else:
        patience_counter += 1
        if patience_counter >= EARLY_STOPPING_PATIENCE:
            print(f"\nEarly stopping triggered after {epoch+1} epochs")
            break

print("\n" + "="*70)
print("TRAINING COMPLETE!")
print("="*70)

# Plot training history
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Training and Validation Loss (Log Scale)')
plt.yscale('log')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nBest validation loss: {best_val_loss:.4f}")
print("Training history saved as 'training_history.png'")
print("Best model saved as 'best_tft_model.pth'")
```

```python

# ==========================================
# 1. Load Libraries
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Set random seeds
torch.manual_seed(42)
np.random.seed(42)

# ==========================================
# 2. Load Dataset from CSV
# ==========================================
df = pd.read_csv('SeoulBikeData.csv', encoding='latin1')
print("CSV Loaded Successfully!")
target_col = 'Rented Bike Count'

# ==========================================
# 3. Feature Engineering Function
# ==========================================
def add_engineered_features(df):
    # --- Fix Date Column ---
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    
    # Extract date parts
    df['Day'] = df['Date'].dt.day
    df['Month'] = df['Date'].dt.month
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
    
    # --- Cyclical Encoding ---
    df['Hour_Sin'] = np.sin(2 * np.pi * df['Hour'] / 24)
    df['Hour_Cos'] = np.cos(2 * np.pi * df['Hour'] / 24)
    df['DayOfWeek_Sin'] = np.sin(2 * np.pi * df['DayOfWeek'] / 7)
    df['DayOfWeek_Cos'] = np.cos(2 * np.pi * df['DayOfWeek'] / 7)
    df['Month_Sin'] = np.sin(2 * np.pi * df['Month'] / 12)
    df['Month_Cos'] = np.cos(2 * np.pi * df['Month'] / 12)
    
    # --- Weather Interaction Features ---
    df['feels_like_temp'] = df['Temperature(°C)'] - (0.1 * df['Humidity(%)'])
    
    # --- Flags (Weekend, Peak Hours) ---
    df['is_weekend'] = df['DayOfWeek'].isin([5, 6]).astype(int)
    df['is_morning_peak'] = df['Hour'].isin([7, 8, 9, 10]).astype(int)
    df['is_evening_peak'] = df['Hour'].isin([17, 18, 19, 20]).astype(int)
    df['is_low_demand_hour'] = df['Hour'].between(0, 5).astype(int)
    
    # --- Binary Conversion ---
    df['Holiday'] = df['Holiday'].map({'No Holiday': 0, 'Holiday': 1})
    df['Functioning Day'] = df['Functioning Day'].map({'Yes': 1, 'No': 0})
    
    # --- Weather binaries ---
    df['Has_Rain'] = (df['Rainfall(mm)'] > 0).astype(int)
    df['Has_Snow'] = (df['Snowfall (cm)'] > 0).astype(int)
    df['Good_Visibility'] = (df['Visibility (10m)'] > 1000).astype(int)
    df['Is_Sunny'] = (df['Solar Radiation (MJ/m2)'] > 0).astype(int)
    
    # --- Target transformation ---
    df['Rented_Bike_Count_Log'] = np.log1p(df['Rented Bike Count'])
    
    # Drop NaNs
    df = df.dropna().reset_index(drop=True)
    
    return df

# ==========================================
# 4. Apply Feature Engineering
# ==========================================
df = add_engineered_features(df)
print("New dataframe shape:", df.shape)
print(df.head())

# ==========================================
# 5. EDA Section (Abbreviated for brevity)
# ==========================================
print("\n" + "="*70)
print("EXPLORATORY DATA ANALYSIS")
print("="*70)

print("\nShape:", df.shape)
print("Date range:", df['Date'].min(), "to", df['Date'].max())
print("\nTarget variable statistics:")
print(df['Rented Bike Count'].describe())

# Quick correlation check
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()
print("\nTop 10 Correlations with Rented Bike Count:")
print(corr['Rented Bike Count'].sort_values(ascending=False).head(10))

# ==========================================
# 6. DATA PREPARATION FOR TFT
# ==========================================
print("\n" + "="*70)
print("DATA PREPARATION FOR TFT MODEL")
print("="*70)

# Encode categorical variables
print("\nEncoding categorical variables...")
label_encoders = {}
for col in ['Seasons']:
    if col in df.columns:
        le = LabelEncoder()
        df[col + '_Encoded'] = le.fit_transform(df[col].astype(str))
        label_encoders[col] = le

# Define feature groups
static_features = ['Seasons_Encoded', 'Holiday', 'Functioning Day']

temporal_numeric_features = [
    'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)',
    'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)',
    'feels_like_temp'
]

temporal_cyclical_features = [
    'Hour_Sin', 'Hour_Cos', 'DayOfWeek_Sin', 'DayOfWeek_Cos',
    'Month_Sin', 'Month_Cos'
]

temporal_binary_features = [
    'is_weekend', 'is_morning_peak', 'is_evening_peak', 'is_low_demand_hour',
    'Has_Rain', 'Has_Snow', 'Good_Visibility', 'Is_Sunny'
]

temporal_features = temporal_numeric_features + temporal_cyclical_features + temporal_binary_features

print(f"Static features: {len(static_features)}")
print(f"Temporal features: {len(temporal_features)}")

# Scale numerical features
print("\nScaling features...")
scaler = StandardScaler()
df[temporal_numeric_features] = scaler.fit_transform(df[temporal_numeric_features])

# Sort by date and hour
df = df.sort_values(['Date', 'Hour']).reset_index(drop=True)
print(f"\nData sorted by Date and Hour")
print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")

# ==========================================
# 7. CREATE TIME SERIES SEQUENCES
# ==========================================
print("\n" + "="*70)
print("CREATING TIME SERIES SEQUENCES")
print("="*70)

# REDUCED LOOKBACK WINDOW
LOOKBACK_WINDOW = 48  # Changed from 72 to 48 (2 days instead of 3)
FORECAST_HORIZON = 24  # 24 hours ahead

def create_sequences(data, lookback, horizon, static_cols, temporal_cols, target_col):
    X_static = []
    X_temporal = []
    y = []
    
    for i in range(lookback, len(data) - horizon + 1):
        static_seq = data[static_cols].iloc[i].values
        temporal_seq = data[temporal_cols].iloc[i-lookback:i].values
        target_seq = data[target_col].iloc[i:i+horizon].values
        
        X_static.append(static_seq)
        X_temporal.append(temporal_seq)
        y.append(target_seq)
    
    return np.array(X_static), np.array(X_temporal), np.array(y)

print(f"Creating sequences with lookback={LOOKBACK_WINDOW}, horizon={FORECAST_HORIZON}...")

X_static, X_temporal, y = create_sequences(
    df,
    LOOKBACK_WINDOW,
    FORECAST_HORIZON,
    static_features,
    temporal_features,
    'Rented_Bike_Count_Log'
)

print(f"X_static shape: {X_static.shape}")
print(f"X_temporal shape: {X_temporal.shape}")
print(f"y shape: {y.shape}")

# ==========================================
# 8. TEMPORAL TRAIN-TEST SPLIT (80/20)
# ==========================================
print("\n" + "="*70)
print("TRAIN-TEST SPLIT (TEMPORAL)")
print("="*70)

split_idx = int(0.8 * len(X_static))

X_static_train = X_static[:split_idx]
X_static_test = X_static[split_idx:]

X_temporal_train = X_temporal[:split_idx]
X_temporal_test = X_temporal[split_idx:]

y_train = y[:split_idx]
y_test = y[split_idx:]

print(f"Train size: {len(X_static_train)} samples")
print(f"Test size: {len(X_static_test)} samples")
print(f"Train percentage: {(len(X_static_train)/len(X_static))*100:.1f}%")
print(f"Test percentage: {(len(X_static_test)/len(X_static))*100:.1f}%")

# Check for distribution shift
print("\n" + "="*70)
print("CHECKING FOR DISTRIBUTION SHIFT")
print("="*70)
print("Train target statistics (original scale):")
print(f"  Mean: {np.expm1(y_train).mean():.2f}")
print(f"  Std: {np.expm1(y_train).std():.2f}")
print(f"  Min: {np.expm1(y_train).min():.2f}")
print(f"  Max: {np.expm1(y_train).max():.2f}")

print("\nTest target statistics (original scale):")
print(f"  Mean: {np.expm1(y_test).mean():.2f}")
print(f"  Std: {np.expm1(y_test).std():.2f}")
print(f"  Min: {np.expm1(y_test).min():.2f}")
print(f"  Max: {np.expm1(y_test).max():.2f}")

# ==========================================
# 9. PYTORCH DATASET AND DATALOADER
# ==========================================
class BikeDataset(Dataset):
    def __init__(self, X_static, X_temporal, y):
        self.X_static = torch.FloatTensor(X_static)
        self.X_temporal = torch.FloatTensor(X_temporal)
        self.y = torch.FloatTensor(y)
    
    def __len__(self):
        return len(self.X_static)
    
    def __getitem__(self, idx):
        return self.X_static[idx], self.X_temporal[idx], self.y[idx]

train_dataset = BikeDataset(X_static_train, X_temporal_train, y_train)
test_dataset = BikeDataset(X_static_test, X_temporal_test, y_test)

BATCH_SIZE = 64
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"\nBatch size: {BATCH_SIZE}")
print(f"Number of training batches: {len(train_loader)}")
print(f"Number of test batches: {len(test_loader)}")

# ==========================================
# 10. IMPROVED TFT MODEL ARCHITECTURE
# ==========================================
print("\n" + "="*70)
print("BUILDING IMPROVED TFT MODEL")
print("="*70)

class VariableSelectionNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout=0.4):
        super().__init__()
        self.hidden_dim = hidden_dim
        
        self.feature_transform = nn.Linear(input_dim, hidden_dim)
        self.bn = nn.BatchNorm1d(hidden_dim)  # Added BatchNorm
        
        self.variable_selection = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.Tanh(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, input_dim),
            nn.Softmax(dim=-1)
        )
        
        self.output_layer = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        transformed = self.feature_transform(x)
        transformed = self.bn(transformed)  # Apply BatchNorm
        weights = self.variable_selection(transformed)
        weighted_features = x * weights
        output = self.output_layer(self.feature_transform(weighted_features))
        return output, weights

class GatedResidualNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout=0.4):
        super().__init__()
        
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.elu = nn.ELU()
        self.fc2 = nn.Linear(hidden_dim, output_dim)
        self.dropout = nn.Dropout(dropout)
        
        self.gate = nn.Sequential(
            nn.Linear(output_dim, output_dim),
            nn.Sigmoid()
        )
        
        if input_dim != output_dim:
            self.skip = nn.Linear(input_dim, output_dim)
        else:
            self.skip = None
        
        self.layer_norm = nn.LayerNorm(output_dim)
    
    def forward(self, x):
        out = self.fc1(x)
        out = self.elu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        
        gate = self.gate(out)
        out = out * gate
        
        if self.skip is not None:
            x = self.skip(x)
        
        out = self.layer_norm(out + x)
        return out

class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads, dropout=0.4):
        super().__init__()
        assert d_model % num_heads == 0
        
        self.d_model = d_model
        self.num_heads = num_heads
        self.d_k = d_model // num_heads
        
        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.W_o = nn.Linear(d_model, d_model)
        
        self.dropout = nn.Dropout(dropout)
        self.layer_norm = nn.LayerNorm(d_model)
    
    def forward(self, query, key, value, mask=None):
        batch_size = query.size(0)
        
        Q = self.W_q(query).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        K = self.W_k(key).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        V = self.W_v(value).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        
        scores = torch.matmul(Q, K.transpose(-2, -1)) / np.sqrt(self.d_k)
        
        if mask is not None:
            scores = scores.masked_fill(mask == 0, -1e9)
        
        attention_weights = torch.softmax(scores, dim=-1)
        attention_weights = self.dropout(attention_weights)
        
        context = torch.matmul(attention_weights, V)
        context = context.transpose(1, 2).contiguous().view(batch_size, -1, self.d_model)
        output = self.W_o(context)
        output = self.layer_norm(output + query)
        
        return output, attention_weights

class EnhancedTemporalFusionTransformer(nn.Module):
    def __init__(self, static_dim, temporal_dim, hidden_dim=64, lstm_layers=1,
                 num_heads=4, dropout=0.4, forecast_horizon=24):
        super().__init__()
        
        self.static_dim = static_dim
        self.temporal_dim = temporal_dim
        self.hidden_dim = hidden_dim
        self.forecast_horizon = forecast_horizon
        
        # Variable Selection Networks (with higher dropout)
        self.static_vsn = VariableSelectionNetwork(
            static_dim, hidden_dim, hidden_dim, dropout
        )
        self.temporal_vsn = VariableSelectionNetwork(
            temporal_dim, hidden_dim, hidden_dim, dropout
        )
        
        # Static Covariate Encoder
        self.static_encoder = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, hidden_dim),
            nn.Dropout(dropout)  # Added extra dropout
        )
        
        # LSTM Layer (reduced to 1 layer)
        self.lstm1 = nn.LSTM(
            hidden_dim, 
            hidden_dim, 
            num_layers=lstm_layers,
            batch_first=True,
            dropout=0  # No dropout since only 1 layer
        )
        
        # Multi-Head Self-Attention (reduced heads)
        self.self_attention = MultiHeadAttention(hidden_dim, num_heads, dropout)
        
        # Gated Residual Networks
        self.grn1 = GatedResidualNetwork(hidden_dim, hidden_dim, hidden_dim, dropout)
        self.grn2 = GatedResidualNetwork(hidden_dim, hidden_dim, hidden_dim, dropout)
        
        # Decoder LSTM
        self.decoder_lstm = nn.LSTM(hidden_dim, hidden_dim, num_layers=1, batch_first=True)
        
        # Output layers with more regularization
        self.output_layer = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim // 2, hidden_dim // 4),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim // 4, 1)
        )
        
        # Quantile outputs
        self.quantile_output = nn.ModuleList([
            nn.Linear(hidden_dim, 1) for _ in range(3)
        ])
    
    def forward(self, static_features, temporal_features):
        batch_size = static_features.size(0)
        seq_len = temporal_features.size(1)
        
        # Variable Selection
        static_encoded, static_weights = self.static_vsn(static_features)
        
        temporal_encoded_list = []
        for t in range(seq_len):
            temp_enc, _ = self.temporal_vsn(temporal_features[:, t, :])
            temporal_encoded_list.append(temp_enc.unsqueeze(1))
        temporal_encoded = torch.cat(temporal_encoded_list, dim=1)
        
        # Static encoding
        static_context = self.static_encoder(static_encoded)
        
        # LSTM processing
        lstm_out1, (h1, c1) = self.lstm1(temporal_encoded)
        
        # Self-Attention
        attention_out, attention_weights = self.self_attention(
            lstm_out1, lstm_out1, lstm_out1
        )
        
        # Gated Residual Networks
        grn_out1 = self.grn1(attention_out)
        
        static_context_expanded = static_context.unsqueeze(1).repeat(1, seq_len, 1)
        combined = grn_out1 + static_context_expanded
        
        grn_out2 = self.grn2(combined)
        
        # Decoder
        context = grn_out2[:, -1, :].unsqueeze(1)
        decoder_input = context.repeat(1, self.forecast_horizon, 1)
        decoder_out, _ = self.decoder_lstm(decoder_input, (h1, c1))
        
        # Predictions
        predictions = self.output_layer(decoder_out).squeeze(-1)
        quantile_preds = [q_layer(decoder_out).squeeze(-1) for q_layer in self.quantile_output]
        
        return predictions, attention_weights, static_weights, quantile_preds

# Initialize model with REDUCED COMPLEXITY
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

model = EnhancedTemporalFusionTransformer(
    static_dim=len(static_features),
    temporal_dim=len(temporal_features),
    hidden_dim=64,      # REDUCED from 128
    lstm_layers=1,      # REDUCED from 2
    num_heads=4,        # REDUCED from 8
    dropout=0.4,        # INCREASED from 0.1
    forecast_horizon=FORECAST_HORIZON
).to(device)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")

# ==========================================
# 11. TRAINING SETUP WITH IMPROVEMENTS
# ==========================================
print("\n" + "="*70)
print("TRAINING CONFIGURATION (IMPROVED)")
print("="*70)

# Try Huber Loss instead of MSE (more robust to outliers)
criterion = nn.HuberLoss(delta=1.0)
print("Loss function: Huber Loss (delta=1.0)")

# LOWER learning rate and HIGHER weight decay
optimizer = optim.Adam(model.parameters(), lr=5e-4, weight_decay=1e-3)
print(f"Learning rate: 5e-4")
print(f"Weight decay: 1e-3")

scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=3  # More aggressive
)

def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    
    progress_bar = tqdm(dataloader, desc="Training")
    for batch_idx, (static, temporal, target) in enumerate(progress_bar):
        static = static.to(device)
        temporal = temporal.to(device)
        target = target.to(device)
        
        optimizer.zero_grad()
        predictions, _, _, _ = model(static, temporal)
        
        loss = criterion(predictions, target)
        loss.backward()
        
        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)  # More aggressive
        optimizer.step()
        
        total_loss += loss.item()
        progress_bar.set_postfix({'loss': loss.item()})
    
    return total_loss / len(dataloader)

def validate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    all_predictions = []
    all_targets = []
    
    with torch.no_grad():
        for static, temporal, target in dataloader:
            static = static.to(device)
            temporal = temporal.to(device)
            target = target.to(device)
            
            predictions, _, _, _ = model(static, temporal)
            loss = criterion(predictions, target)
            
            total_loss += loss.item()
            all_predictions.append(predictions.cpu().numpy())
            all_targets.append(target.cpu().numpy())
    
    all_predictions = np.concatenate(all_predictions, axis=0)
    all_targets = np.concatenate(all_targets, axis=0)
    
    return total_loss / len(dataloader), all_predictions, all_targets

# ==========================================
# 12. TRAINING LOOP WITH IMPROVEMENTS
# ==========================================
print("\n" + "="*70)
print("STARTING TRAINING (IMPROVED)")
print("="*70)

NUM_EPOCHS = 100  # Increased to give model more time
EARLY_STOPPING_PATIENCE = 15  # Increased patience

train_losses = []
val_losses = []
train_maes = []
val_maes = []
train_rmses = []
val_rmses = []
best_val_loss = float('inf')
best_val_rmse = float('inf')
patience_counter = 0
current_lr = optimizer.param_groups[0]['lr']

print(f"Training for up to {NUM_EPOCHS} epochs")
print(f"Early stopping patience: {EARLY_STOPPING_PATIENCE}")
print(f"Initial learning rate: {current_lr:.2e}\n")

for epoch in range(NUM_EPOCHS):
    print(f"\nEpoch {epoch+1}/{NUM_EPOCHS}")
    print("-" * 70)
    
    # Train
    train_loss = train_epoch(model, train_loader, criterion, optimizer, device)
    train_losses.append(train_loss)
    
    # Validate
    val_loss, val_preds, val_targets = validate(model, test_loader, criterion, device)
    val_losses.append(val_loss)
    
    # Also get training predictions for monitoring
    _, train_preds, train_targets = validate(model, train_loader, criterion, device)
    
    # Check if learning rate changed
    old_lr = current_lr
    scheduler.step(val_loss)
    current_lr = optimizer.param_groups[0]['lr']
    if current_lr != old_lr:
        print(f"⚠️  Learning rate reduced: {old_lr:.2e} → {current_lr:.2e}")
    
    # Calculate metrics (convert back from log scale)
    train_preds_original = np.expm1(train_preds)
    train_targets_original = np.expm1(train_targets)
    val_preds_original = np.expm1(val_preds)
    val_targets_original = np.expm1(val_targets)
    
    # Training metrics
    train_mae = np.mean(np.abs(train_preds_original - train_targets_original))
    train_rmse = np.sqrt(np.mean((train_preds_original - train_targets_original) ** 2))
    train_mape = np.mean(np.abs((train_targets_original - train_preds_original) / (train_targets_original + 1))) * 100
    
    # Validation metrics
    val_mae = np.mean(np.abs(val_preds_original - val_targets_original))
    val_rmse = np.sqrt(np.mean((val_preds_original - val_targets_original) ** 2))
    val_mape = np.mean(np.abs((val_targets_original - val_preds_original) / (val_targets_original + 1))) * 100
    
    train_maes.append(train_mae)
    val_maes.append(val_mae)
    train_rmses.append(train_rmse)
    val_rmses.append(val_rmse)
    
    # Print metrics
    print(f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
    print(f"Train: MAE={train_mae:.2f}, RMSE={train_rmse:.2f}, MAPE={train_mape:.2f}%")
    print(f"Val:   MAE={val_mae:.2f}, RMSE={val_rmse:.2f}, MAPE={val_mape:.2f}%")
    print(f"Learning Rate: {current_lr:.2e}")
    
    # Calculate overfitting ratio
    overfit_ratio = val_loss / (train_loss + 1e-8)
    print(f"Validation/Train Loss Ratio: {overfit_ratio:.2f}x")
    
    # Early stopping based on validation loss
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_val_rmse = val_rmse
        patience_counter = 0
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_loss': val_loss,
            'val_mae': val_mae,
            'val_rmse': val_rmse,
            'val_mape': val_mape,
            'train_loss': train_loss,
            'train_mae': train_mae,
            'train_rmse': train_rmse
        }, 'best_tft_model.pth')
        print("✅ Best model saved!")
    else:
        patience_counter += 1
        print(f"⏳ Patience: {patience_counter}/{EARLY_STOPPING_PATIENCE}")
        if patience_counter >= EARLY_STOPPING_PATIENCE:
            print(f"\n🛑 Early stopping triggered after {epoch+1} epochs")
            break

print("\n" + "="*70)
print("TRAINING COMPLETE!")
print("="*70)
print(f"Best Validation Loss: {best_val_loss:.4f}")
print(f"Best Validation RMSE: {best_val_rmse:.2f}")

# ==========================================
# 13. VISUALIZATION AND ANALYSIS
# ==========================================
print("\n" + "="*70)
print("GENERATING VISUALIZATIONS")
print("="*70)

# Create comprehensive visualization
fig = plt.figure(figsize=(18, 12))

# Plot 1: Loss curves
ax1 = plt.subplot(3, 3, 1)
plt.plot(train_losses, label='Train Loss', linewidth=2)
plt.plot(val_losses, label='Val Loss', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Loss (Huber)')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(alpha=0.3)

# Plot 2: Loss curves (log scale)
ax2 = plt.subplot(3, 3, 2)
plt.plot(train_losses, label='Train Loss', linewidth=2)
plt.plot(val_losses, label='Val Loss', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Loss (Huber)')
plt.title('Loss (Log Scale)')
plt.yscale('log')
plt.legend()
plt.grid(alpha=0.3)

# Plot 3: Overfitting ratio
ax3 = plt.subplot(3, 3, 3)
overfit_ratios = [v/(t+1e-8) for v, t in zip(val_losses, train_losses)]
plt.plot(overfit_ratios, linewidth=2, color='red')
plt.axhline(y=1.0, color='green', linestyle='--', label='No Overfitting')
plt.axhline(y=2.0, color='orange', linestyle='--', label='2x Overfitting')
plt.xlabel('Epoch')
plt.ylabel('Val/Train Ratio')
plt.title('Overfitting Ratio Over Time')
plt.legend()
plt.grid(alpha=0.3)

# Plot 4: MAE curves
ax4 = plt.subplot(3, 3, 4)
plt.plot(train_maes, label='Train MAE', linewidth=2)
plt.plot(val_maes, label='Val MAE', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('MAE (bikes)')
plt.title('Mean Absolute Error')
plt.legend()
plt.grid(alpha=0.3)

# Plot 5: RMSE curves
ax5 = plt.subplot(3, 3, 5)
plt.plot(train_rmses, label='Train RMSE', linewidth=2)
plt.plot(val_rmses, label='Val RMSE', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('RMSE (bikes)')
plt.title('Root Mean Squared Error')
plt.legend()
plt.grid(alpha=0.3)

# Plot 6: Learning rate over time
ax6 = plt.subplot(3, 3, 6)
# Reconstruct LR history (approximate)
lr_history = [5e-4] * len(train_losses)
plt.plot(lr_history, linewidth=2, color='purple')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedule')
plt.yscale('log')
plt.grid(alpha=0.3)

# Load best model for final predictions
print("\nLoading best model for final evaluation...")
checkpoint = torch.load('best_tft_model.pth')
model.load_state_dict(checkpoint['model_state_dict'])
print(f"Best model from epoch {checkpoint['epoch']+1} loaded")

# Get final predictions
_, final_val_preds, final_val_targets = validate(model, test_loader, criterion, device)
_, final_train_preds, final_train_targets = validate(model, train_loader, criterion, device)

# Convert to original scale
final_val_preds_orig = np.expm1(final_val_preds)
final_val_targets_orig = np.expm1(final_val_targets)
final_train_preds_orig = np.expm1(final_train_preds)
final_train_targets_orig = np.expm1(final_train_targets)

# Plot 7: Validation predictions vs actual (scatter)
ax7 = plt.subplot(3, 3, 7)
# Take only first timestep for clearer visualization
val_pred_t0 = final_val_preds_orig[:, 0]
val_target_t0 = final_val_targets_orig[:, 0]
plt.scatter(val_target_t0, val_pred_t0, alpha=0.3, s=10)
plt.plot([0, val_target_t0.max()], [0, val_target_t0.max()], 'r--', linewidth=2, label='Perfect Prediction')
plt.xlabel('Actual Bike Count')
plt.ylabel('Predicted Bike Count')
plt.title('Validation: Predictions vs Actual (t+1)')
plt.legend()
plt.grid(alpha=0.3)

# Plot 8: Residuals distribution
ax8 = plt.subplot(3, 3, 8)
residuals = val_pred_t0 - val_target_t0
plt.hist(residuals, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(x=0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Prediction Error (bikes)')
plt.ylabel('Frequency')
plt.title(f'Residuals Distribution\nMean: {residuals.mean():.2f}, Std: {residuals.std():.2f}')
plt.grid(alpha=0.3)

# Plot 9: Time series comparison (sample)
ax9 = plt.subplot(3, 3, 9)
# Show predictions for a sample period
sample_idx = 100
sample_size = 50
if sample_idx + sample_size < len(final_val_preds_orig):
    sample_preds = final_val_preds_orig[sample_idx:sample_idx+sample_size, 0]
    sample_targets = final_val_targets_orig[sample_idx:sample_idx+sample_size, 0]
    x_axis = range(len(sample_preds))
    plt.plot(x_axis, sample_targets, label='Actual', linewidth=2, marker='o', markersize=3)
    plt.plot(x_axis, sample_preds, label='Predicted', linewidth=2, marker='s', markersize=3)
    plt.xlabel('Time Step')
    plt.ylabel('Bike Count')
    plt.title('Sample Predictions vs Actual')
    plt.legend()
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('comprehensive_training_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Comprehensive visualization saved as 'comprehensive_training_analysis.png'")

# ==========================================
# 14. FINAL EVALUATION METRICS
# ==========================================
print("\n" + "="*70)
print("FINAL EVALUATION METRICS")
print("="*70)

# Calculate comprehensive metrics
def calculate_metrics(predictions, targets):
    """Calculate all evaluation metrics"""
    mae = np.mean(np.abs(predictions - targets))
    mse = np.mean((predictions - targets) ** 2)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((targets - predictions) / (targets + 1))) * 100
    
    # R-squared
    ss_res = np.sum((targets - predictions) ** 2)
    ss_tot = np.sum((targets - np.mean(targets)) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    
    return {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'MAPE': mape,
        'R2': r2
    }

# Evaluate for each forecast horizon
print("\n📊 METRICS BY FORECAST HORIZON:")
print("-" * 70)

for horizon_idx in range(min(24, FORECAST_HORIZON)):
    horizon_preds = final_val_preds_orig[:, horizon_idx]
    horizon_targets = final_val_targets_orig[:, horizon_idx]
    metrics = calculate_metrics(horizon_preds, horizon_targets)
    
    if horizon_idx < 3 or horizon_idx % 6 == 0 or horizon_idx == 23:  # Print selected horizons
        print(f"\nHorizon t+{horizon_idx+1}:")
        print(f"  MAE:  {metrics['MAE']:.2f} bikes")
        print(f"  RMSE: {metrics['RMSE']:.2f} bikes")
        print(f"  MAPE: {metrics['MAPE']:.2f}%")
        print(f"  R²:   {metrics['R2']:.4f}")

# Overall metrics (averaged across all horizons)
print("\n" + "="*70)
print("📈 OVERALL PERFORMANCE (AVERAGED ACROSS ALL HORIZONS)")
print("="*70)

overall_metrics = calculate_metrics(final_val_preds_orig.flatten(), final_val_targets_orig.flatten())
train_metrics = calculate_metrics(final_train_preds_orig.flatten(), final_train_targets_orig.flatten())

print("\n🔵 TRAINING SET:")
print(f"  MAE:  {train_metrics['MAE']:.2f} bikes")
print(f"  RMSE: {train_metrics['RMSE']:.2f} bikes")
print(f"  MAPE: {train_metrics['MAPE']:.2f}%")
print(f"  R²:   {train_metrics['R2']:.4f}")

print("\n🟢 VALIDATION SET:")
print(f"  MAE:  {overall_metrics['MAE']:.2f} bikes")
print(f"  RMSE: {overall_metrics['RMSE']:.2f} bikes")
print(f"  MAPE: {overall_metrics['MAPE']:.2f}%")
print(f"  R²:   {overall_metrics['R2']:.4f}")

# Generalization gap
print("\n⚖️  GENERALIZATION GAP:")
print(f"  MAE Gap:  {overall_metrics['MAE'] - train_metrics['MAE']:.2f} bikes ({((overall_metrics['MAE']/train_metrics['MAE']-1)*100):.1f}%)")
print(f"  RMSE Gap: {overall_metrics['RMSE'] - train_metrics['RMSE']:.2f} bikes ({((overall_metrics['RMSE']/train_metrics['RMSE']-1)*100):.1f}%)")
print(f"  R² Gap:   {train_metrics['R2'] - overall_metrics['R2']:.4f}")

# ==========================================
# 15. DETAILED PREDICTIONS VISUALIZATION
# ==========================================
print("\n" + "="*70)
print("CREATING DETAILED PREDICTIONS VISUALIZATION")
print("="*70)

fig = plt.figure(figsize=(20, 12))

# Plot 1: Multi-horizon predictions for a sample
ax1 = plt.subplot(3, 2, 1)
sample_idx = 200
if sample_idx < len(final_val_preds_orig):
    horizons = range(1, min(25, FORECAST_HORIZON+1))
    actual_values = final_val_targets_orig[sample_idx, :24]
    pred_values = final_val_preds_orig[sample_idx, :24]
    
    plt.plot(horizons, actual_values, 'o-', label='Actual', linewidth=2, markersize=6)
    plt.plot(horizons, pred_values, 's-', label='Predicted', linewidth=2, markersize=6)
    plt.fill_between(horizons, actual_values, pred_values, alpha=0.2)
    plt.xlabel('Forecast Horizon (hours ahead)')
    plt.ylabel('Bike Count')
    plt.title(f'Multi-Horizon Forecast (Sample #{sample_idx})')
    plt.legend()
    plt.grid(alpha=0.3)

# Plot 2: Error by forecast horizon
ax2 = plt.subplot(3, 2, 2)
horizon_maes = []
horizon_rmses = []
for h in range(min(24, FORECAST_HORIZON)):
    h_mae = np.mean(np.abs(final_val_preds_orig[:, h] - final_val_targets_orig[:, h]))
    h_rmse = np.sqrt(np.mean((final_val_preds_orig[:, h] - final_val_targets_orig[:, h])**2))
    horizon_maes.append(h_mae)
    horizon_rmses.append(h_rmse)

x_horizons = range(1, len(horizon_maes)+1)
plt.plot(x_horizons, horizon_maes, 'o-', label='MAE', linewidth=2, markersize=4)
plt.plot(x_horizons, horizon_rmses, 's-', label='RMSE', linewidth=2, markersize=4)
plt.xlabel('Forecast Horizon (hours ahead)')
plt.ylabel('Error (bikes)')
plt.title('Prediction Error by Forecast Horizon')
plt.legend()
plt.grid(alpha=0.3)

# Plot 3: Error distribution by time of day
ax3 = plt.subplot(3, 2, 3)
# Map predictions back to hour of day (approximate)
hours = np.tile(np.arange(24), len(final_val_preds_orig) // 24 + 1)[:len(final_val_preds_orig)]
hourly_errors = []
hourly_stds = []
for h in range(24):
    hour_mask = hours == h
    if hour_mask.sum() > 0:
        hour_error = np.abs(final_val_preds_orig[hour_mask, 0] - final_val_targets_orig[hour_mask, 0])
        hourly_errors.append(hour_error.mean())
        hourly_stds.append(hour_error.std())
    else:
        hourly_errors.append(0)
        hourly_stds.append(0)

plt.bar(range(24), hourly_errors, alpha=0.7, edgecolor='black')
plt.xlabel('Hour of Day')
plt.ylabel('Mean Absolute Error (bikes)')
plt.title('Prediction Error by Hour of Day (t+1)')
plt.grid(alpha=0.3, axis='y')
plt.xticks(range(0, 24, 3))

# Plot 4: Prediction intervals (with quantiles if available)
ax4 = plt.subplot(3, 2, 4)
sample_start = 50
sample_length = 100
if sample_start + sample_length < len(final_val_preds_orig):
    x_range = range(sample_length)
    actual = final_val_targets_orig[sample_start:sample_start+sample_length, 0]
    pred = final_val_preds_orig[sample_start:sample_start+sample_length, 0]
    
    plt.plot(x_range, actual, label='Actual', linewidth=2, alpha=0.8)
    plt.plot(x_range, pred, label='Predicted', linewidth=2, alpha=0.8)
    
    # Add confidence bands (using prediction std as proxy)
    pred_std = np.std(pred - actual)
    plt.fill_between(x_range, pred - 1.96*pred_std, pred + 1.96*pred_std, 
                     alpha=0.2, label='95% Confidence')
    
    plt.xlabel('Time Step')
    plt.ylabel('Bike Count')
    plt.title('Predictions with Confidence Interval')
    plt.legend()
    plt.grid(alpha=0.3)

# Plot 5: Error heatmap by day and hour
ax5 = plt.subplot(3, 2, 5)
# Create day x hour error matrix
days = min(30, len(final_val_preds_orig) // 24)
if days > 0:
    error_matrix = np.zeros((days, 24))
    for d in range(days):
        for h in range(24):
            idx = d * 24 + h
            if idx < len(final_val_preds_orig):
                error_matrix[d, h] = abs(final_val_preds_orig[idx, 0] - final_val_targets_orig[idx, 0])
    
    im = plt.imshow(error_matrix, aspect='auto', cmap='YlOrRd', interpolation='nearest')
    plt.colorbar(im, label='Absolute Error (bikes)')
    plt.xlabel('Hour of Day')
    plt.ylabel('Day')
    plt.title('Prediction Error Heatmap (First 30 Days)')
    plt.xticks(range(0, 24, 3))

# Plot 6: Cumulative error distribution
ax6 = plt.subplot(3, 2, 6)
errors = np.abs(final_val_preds_orig[:, 0] - final_val_targets_orig[:, 0])
sorted_errors = np.sort(errors)
cumulative = np.arange(1, len(sorted_errors) + 1) / len(sorted_errors) * 100

plt.plot(sorted_errors, cumulative, linewidth=2)
plt.axhline(y=50, color='r', linestyle='--', alpha=0.7, label='50th percentile')
plt.axhline(y=90, color='orange', linestyle='--', alpha=0.7, label='90th percentile')
plt.axhline(y=95, color='red', linestyle='--', alpha=0.7, label='95th percentile')
plt.xlabel('Absolute Error (bikes)')
plt.ylabel('Cumulative Percentage (%)')
plt.title('Cumulative Error Distribution (t+1)')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('detailed_predictions_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Detailed predictions visualization saved as 'detailed_predictions_analysis.png'")

# ==========================================
# 16. ATTENTION WEIGHTS ANALYSIS
# ==========================================
print("\n" + "="*70)
print("ANALYZING ATTENTION WEIGHTS")
print("="*70)

# Get attention weights for a sample
model.eval()
with torch.no_grad():
    sample_static = torch.FloatTensor(X_static_test[:5]).to(device)
    sample_temporal = torch.FloatTensor(X_temporal_test[:5]).to(device)
    _, attention_weights, static_weights, _ = model(sample_static, sample_temporal)
    
    # Average across samples and heads
    avg_attention = attention_weights.mean(dim=0).mean(dim=0).cpu().numpy()  # (seq_len, seq_len)
    avg_static_weights = static_weights.mean(dim=0).cpu().numpy()  # (static_features,)

# Plot attention patterns
fig = plt.figure(figsize=(16, 6))

# Temporal attention heatmap
ax1 = plt.subplot(1, 3, 1)
im1 = plt.imshow(avg_attention, cmap='viridis', aspect='auto')
plt.colorbar(im1, label='Attention Weight')
plt.xlabel('Key Position')
plt.ylabel('Query Position')
plt.title('Average Temporal Self-Attention Pattern')

# Static feature importance
ax2 = plt.subplot(1, 3, 2)
plt.barh(static_features, avg_static_weights)
plt.xlabel('Average Attention Weight')
plt.title('Static Feature Importance')
plt.grid(alpha=0.3, axis='x')

# Attention by time lag
ax3 = plt.subplot(1, 3, 3)
# Average attention as function of time lag
seq_len = avg_attention.shape[0]
lag_attention = []
for lag in range(seq_len):
    # Diagonal elements at different lags
    diag_vals = []
    for i in range(seq_len - lag):
        diag_vals.append(avg_attention[i + lag, i])
    if diag_vals:
        lag_attention.append(np.mean(diag_vals))

plt.plot(range(len(lag_attention)), lag_attention, 'o-', linewidth=2, markersize=6)
plt.xlabel('Time Lag (hours)')
plt.ylabel('Average Attention Weight')
plt.title('Attention vs Time Lag')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('attention_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Attention analysis saved as 'attention_analysis.png'")

# ==========================================
# 17. COMPARISON WITH BASELINE
# ==========================================
print("\n" + "="*70)
print("BASELINE COMPARISON")
print("="*70)

# Simple baseline: persistence model (use last known value)
baseline_preds = np.repeat(
    np.expm1(X_temporal_test[:, -1, temporal_features.index('Temperature(°C)')]).reshape(-1, 1),
    FORECAST_HORIZON,
    axis=1
)

# For a better baseline, use last hour's bike count from the data
# This requires accessing the original data
baseline_mae = np.mean(np.abs(baseline_preds - final_val_targets_orig))
baseline_rmse = np.sqrt(np.mean((baseline_preds - final_val_targets_orig)**2))

print("\n📊 PERFORMANCE COMPARISON:")
print("-" * 70)
print(f"{'Model':<30} {'MAE':<15} {'RMSE':<15} {'Improvement'}")
print("-" * 70)
print(f"{'Persistence Baseline':<30} {baseline_mae:<15.2f} {baseline_rmse:<15.2f} -")
print(f"{'Enhanced TFT (Ours)':<30} {overall_metrics['MAE']:<15.2f} {overall_metrics['RMSE']:<15.2f} "
      f"{((baseline_rmse - overall_metrics['RMSE'])/baseline_rmse*100):.1f}%")

# ==========================================
# 18. SAVE FINAL RESULTS
# ==========================================
print("\n" + "="*70)
print("SAVING FINAL RESULTS")
print("="*70)

# Save predictions
results_df = pd.DataFrame({
    'actual_mean': final_val_targets_orig.mean(axis=1),
    'predicted_mean': final_val_preds_orig.mean(axis=1),
    'error': np.abs(final_val_targets_orig.mean(axis=1) - final_val_preds_orig.mean(axis=1))
})
results_df.to_csv('predictions_results.csv', index=False)
print("✅ Predictions saved to 'predictions_results.csv'")

# Save metrics summary
metrics_summary = {
    'model_config': {
        'hidden_dim': 64,
        'lstm_layers': 1,
        'num_heads': 4,
        'dropout': 0.4,
        'lookback_window': LOOKBACK_WINDOW,
        'forecast_horizon': FORECAST_HORIZON
    },
    'training': {
        'epochs_trained': len(train_losses),
        'best_epoch': checkpoint['epoch'] + 1,
        'final_lr': current_lr
    },
    'performance': {
        'train': train_metrics,
        'validation': overall_metrics,
        'best_val_loss': float(best_val_loss),
        'best_val_rmse': float(best_val_rmse)
    }
}

import json
with open('model_metrics_summary.json', 'w') as f:
    json.dump(metrics_summary, f, indent=4)
print("✅ Metrics summary saved to 'model_metrics_summary.json'")

print("\n" + "="*70)
print("🎉 ALL DONE!")
print("="*70)
print("\n📁 Generated Files:")
print("  1. best_tft_model.pth - Best model weights")
print("  2. comprehensive_training_analysis.png - Training curves")
print("  3. detailed_predictions_analysis.png - Prediction analysis")
print("  4. attention_analysis.png - Attention weights visualization")
print("  5. predictions_results.csv - Prediction results")
print("  6. model_metrics_summary.json - Complete metrics")
print("\n✨ Model training and evaluation complete!")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler

# -------------------- 1. Load and Feature Engineering --------------------

df = pd.read_csv('SeoulBikeData.csv', encoding='latin1')
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df = df.sort_values(['Date', 'Hour']).reset_index(drop=True)
target_col = 'Rented Bike Count'

def add_engineered_features(df, max_lag=3):
    df['Day'] = df['Date'].dt.day
    df['Month'] = df['Date'].dt.month
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
    df['feels_like_temp'] = df['Temperature(°C)'] - (0.1 * df['Humidity(%)'])
    df['is_weekend'] = df['DayOfWeek'].isin([5,6]).astype(int)
    df['is_morning_peak'] = df['Hour'].isin([7,8,9,10]).astype(int)
    df['is_evening_peak'] = df['Hour'].isin([17,18,19,20]).astype(int)
    df['is_low_demand_hour'] = df['Hour'].between(0,5).astype(int)
    df['Holiday'] = df['Holiday'].map({'No Holiday':0, 'Holiday':1})
    df['Functioning Day'] = df['Functioning Day'].map({'Yes':1, 'No':0})
    for lag in range(1, max_lag+1):
        df[f'lag_{lag}'] = df[target_col].shift(lag)
    if 'Seasons' in df.columns:
        df = pd.get_dummies(df, columns=['Seasons'], drop_first=True)
    df = df.dropna().reset_index(drop=True)
    return df

df = add_engineered_features(df)
df = df.fillna(0)

feature_cols = [
    'Hour', 'Day', 'Month', 'DayOfWeek', 'WeekOfYear',
    'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)',
    'Visibility (10m)', 'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 
    'Rainfall(mm)', 'Snowfall (cm)', 'Holiday', 'Functioning Day',
    'feels_like_temp', 'is_weekend', 'is_morning_peak', 'is_evening_peak', 'is_low_demand_hour',
    'lag_1', 'lag_2', 'lag_3'
]
feature_cols += [col for col in df.columns if col.startswith('Seasons_')]

# -------------------- 2. Create windowed sequences --------------------

window_size = 24  # input sequence length
X_win, y_win = [], []

for i in range(window_size, len(df)):
    X_win.append(df[feature_cols].iloc[i-window_size:i].values)
    y_win.append(df[target_col].iloc[i])

X_win = np.array(X_win, dtype=np.float32)
y_win = np.array(y_win, dtype=np.float32)

# -------------------- 3. Chronological Train/Val/Test Split --------------------

unique_dates = pd.Series(df['Date'].iloc[window_size:]).dt.date.unique()
total_days = len(unique_dates)
train_days = int(total_days * 0.7)
val_days = int(total_days * 0.15)
test_days = total_days - train_days - val_days

train_dates = unique_dates[:train_days]
val_dates = unique_dates[train_days:train_days+val_days]
test_dates = unique_dates[train_days+val_days:]

dates_series = pd.Series(df['Date'].iloc[window_size:]).dt.date

train_idx = dates_series.isin(train_dates)
val_idx = dates_series.isin(val_dates)
test_idx = dates_series.isin(test_dates)

X_train = X_win[train_idx.values]
y_train = y_win[train_idx.values]
X_val = X_win[val_idx.values]
y_val = y_win[val_idx.values]
X_test = X_win[test_idx.values]
y_test = y_win[test_idx.values]

# -------------------- 4. Scaling --------------------

scaler_X = StandardScaler()
scaler_y = StandardScaler()

# Flatten X for scaler fitting
X_train_flat = X_train.reshape(-1, X_train.shape[-1])
X_val_flat = X_val.reshape(-1, X_val.shape[-1])
X_test_flat = X_test.reshape(-1, X_test.shape[-1])

X_train_scaled_flat = scaler_X.fit_transform(X_train_flat)
X_val_scaled_flat = scaler_X.transform(X_val_flat)
X_test_scaled_flat = scaler_X.transform(X_test_flat)

# Reshape back to 3D
X_train_scaled = X_train_scaled_flat.reshape(X_train.shape)
X_val_scaled = X_val_scaled_flat.reshape(X_val.shape)
X_test_scaled = X_test_scaled_flat.reshape(X_test.shape)

y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1,1)).flatten()
y_val_scaled = scaler_y.transform(y_val.reshape(-1,1)).flatten()
y_test_scaled = scaler_y.transform(y_test.reshape(-1,1)).flatten()

# Convert to torch tensors and dataloaders
train_ds = TensorDataset(torch.from_numpy(X_train_scaled), torch.from_numpy(y_train_scaled).unsqueeze(1))
val_ds = TensorDataset(torch.from_numpy(X_val_scaled), torch.from_numpy(y_val_scaled).unsqueeze(1))
test_ds = TensorDataset(torch.from_numpy(X_test_scaled), torch.from_numpy(y_test_scaled).unsqueeze(1))

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=32)
test_loader = DataLoader(test_ds, batch_size=32)

# -------------------- 5. TFT-like Model --------------------

class GatedResidualNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim=None, dropout=0.1):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.elu = nn.ELU()
        self.fc2 = nn.Linear(hidden_dim, output_dim or input_dim)
        self.dropout = nn.Dropout(dropout)
        self.gate = nn.Linear(output_dim or input_dim, output_dim or input_dim)
        self.skip = nn.Linear(input_dim, output_dim or input_dim) if output_dim != input_dim else None
        self.norm = nn.LayerNorm(output_dim or input_dim)
    def forward(self, x):
        residual = x
        x = self.elu(self.fc1(x))
        x = self.dropout(self.fc2(x))
        gate = torch.sigmoid(self.gate(x))
        if self.skip:
            residual = self.skip(residual)
        x = gate * x + (1 - gate) * residual
        return self.norm(x)

class TemporalFusionTransformer(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, dropout=0.1, output_dim=1):
        super().__init__()
        self.input_linear = nn.Linear(input_dim, hidden_dim)
        self.grn1 = GatedResidualNetwork(hidden_dim, hidden_dim, dropout=dropout)
        self.lstm = nn.LSTM(hidden_dim, hidden_dim, batch_first=True)
        self.attention = nn.MultiheadAttention(hidden_dim, num_heads=4, dropout=dropout, batch_first=True)
        self.grn2 = GatedResidualNetwork(hidden_dim, hidden_dim, dropout=dropout)
        self.output_layer = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = self.input_linear(x)
        x = self.grn1(x)
        lstm_out, _ = self.lstm(x)
        attn_out, _ = self.attention(lstm_out, lstm_out, lstm_out)
        combined = lstm_out + attn_out
        combined = self.grn2(combined)
        pooled = combined.mean(dim=1)
        out = self.output_layer(pooled)
        return out

# Instantiate model
model = TemporalFusionTransformer(input_dim=X_train.shape[2], hidden_dim=128)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# -------------------- 6. Training --------------------

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=2e-4)
epochs = 70

best_val_loss = float('inf')

for epoch in range(epochs):
    model.train()
    train_losses = []
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())

    model.eval()
    val_losses = []
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            preds = model(xb)
            val_losses.append(criterion(preds, yb).item())

    avg_train_loss = np.mean(train_losses)
    avg_val_loss = np.mean(val_losses)
    print(f"Epoch {epoch+1}/{epochs} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")

    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        torch.save(model.state_dict(), "best_tft_model.pth")

# -------------------- 7. Evaluation --------------------

model.load_state_dict(torch.load("best_tft_model.pth"))
model.eval()

y_true_all = []
y_pred_all = []

with torch.no_grad():
    for xb, yb in test_loader:
        xb, yb = xb.to(device), yb.to(device)
        preds = model(xb)
        y_true_all.append(yb.cpu().numpy())
        y_pred_all.append(preds.cpu().numpy())

y_true_all = np.concatenate(y_true_all, axis=0)
y_pred_all = np.concatenate(y_pred_all, axis=0)

y_true_all = scaler_y.inverse_transform(y_true_all)
y_pred_all = scaler_y.inverse_transform(y_pred_all)

from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y_true_all, y_pred_all)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true_all, y_pred_all)
safe_mask = y_true_all.flatten() > 10
mape = np.mean(np.abs((y_true_all[safe_mask] - y_pred_all[safe_mask]) / y_true_all[safe_mask])) * 100
smape = np.mean(2 * np.abs(y_pred_all - y_true_all) / (np.abs(y_true_all) + np.abs(y_pred_all) + 1e-8)) * 100

print(f"Test MSE: {mse:.4f}")
print(f"Test RMSE: {rmse:.4f}")
print(f"Test MAE: {mae:.4f}")
print(f"Test MAPE: {mape:.2f}%")
print(f"Test SMAPE: {smape:.2f}%")
```

```python
# ============================================================
# 1. Import Libraries
# ============================================================
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler

# ============================================================
# 2. Load Dataset
# ============================================================
df = pd.read_csv("SeoulBikeData.csv", encoding="latin1")
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df = df.sort_values(["Date", "Hour"]).reset_index(drop=True)

target_col = "Rented Bike Count"

# ============================================================
# 3. Feature Engineering
# ============================================================
def add_engineered_features(df, max_lag=3):

    df["Day"] = df["Date"].dt.day
    df["Month"] = df["Date"].dt.month
    df["DayOfWeek"] = df["Date"].dt.dayofweek
    df["WeekOfYear"] = df["Date"].dt.isocalendar().week.astype(int)

    df["feels_like_temp"] = df["Temperature(°C)"] - (0.1 * df["Humidity(%)"])
    df["is_weekend"] = df["DayOfWeek"].isin([5, 6]).astype(int)
    df["is_morning_peak"] = df["Hour"].isin([7, 8, 9, 10]).astype(int)
    df["is_evening_peak"] = df["Hour"].isin([17, 18, 19, 20]).astype(int)
    df["is_low_demand_hour"] = df["Hour"].between(0, 5).astype(int)

    df["Holiday"] = df["Holiday"].map({"No Holiday": 0, "Holiday": 1})
    df["Functioning Day"] = df["Functioning Day"].map({"Yes": 1, "No": 0})

    # Lag features (very important for TFT-like behaviour)
    for lag in range(1, max_lag + 1):
        df[f"lag_{lag}"] = df[target_col].shift(lag)

    # One-hot encode seasons if present
    if "Seasons" in df.columns:
        df = pd.get_dummies(df, columns=["Seasons"], drop_first=True)

    return df.dropna().reset_index(drop=True)


df = add_engineered_features(df)

# ============================================================
# 4. Select Feature Columns
# ============================================================
feature_cols = [
    'Hour','Day','Month','DayOfWeek','WeekOfYear',
    'Temperature(°C)','Humidity(%)','Wind speed (m/s)',
    'Visibility (10m)','Dew point temperature(°C)','Solar Radiation (MJ/m2)',
    'Rainfall(mm)','Snowfall (cm)','Holiday','Functioning Day',
    'feels_like_temp','is_weekend','is_morning_peak','is_evening_peak',
    'is_low_demand_hour','lag_1','lag_2','lag_3'
]

feature_cols += [c for c in df.columns if c.startswith("Seasons_")]

# ============================================================
# 5. Create Windowed Sequences
# ============================================================
window_size = 24

X_seq, y_seq, date_seq = [], [], []

for i in range(window_size, len(df)):
    X_seq.append(df[feature_cols].iloc[i-window_size:i].values)
    y_seq.append(df[target_col].iloc[i])
    date_seq.append(df["Date"].iloc[i])

X_seq = np.array(X_seq, dtype=np.float32)
y_seq = np.array(y_seq, dtype=np.float32)
date_seq = np.array(date_seq)

# ============================================================
# 6. Chronological 80/10/10 Split by Days
# ============================================================
date_arr = np.array([d.date() for d in date_seq])
unique_days = np.unique(date_arr)

n_train = int(len(unique_days) * 0.8)
n_val = int(len(unique_days) * 0.1)

train_days = unique_days[:n_train]
val_days = unique_days[n_train:n_train+n_val]
test_days = unique_days[n_train+n_val:]

train_mask = np.isin(date_arr, train_days)
val_mask = np.isin(date_arr, val_days)
test_mask = np.isin(date_arr, test_days)

X_train = X_seq[train_mask]
X_val = X_seq[val_mask]
X_test = X_seq[test_mask]

y_train = y_seq[train_mask]
y_val = y_seq[val_mask]
y_test = y_seq[test_mask]

print(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")

# ============================================================
# 7. Scaling
# ============================================================
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_flat = X_train.reshape(-1, X_train.shape[-1])
X_val_flat = X_val.reshape(-1, X_val.shape[-1])
X_test_flat = X_test.reshape(-1, X_test.shape[-1])

X_train_scaled_flat = scaler_X.fit_transform(X_train_flat)
X_val_scaled_flat = scaler_X.transform(X_val_flat)
X_test_scaled_flat = scaler_X.transform(X_test_flat)

X_train_scaled = X_train_scaled_flat.reshape(X_train.shape)
X_val_scaled = X_val_scaled_flat.reshape(X_val.shape)
X_test_scaled = X_test_scaled_flat.reshape(X_test.shape)

y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1,1)).flatten()
y_val_scaled = scaler_y.transform(y_val.reshape(-1,1)).flatten()
y_test_scaled = scaler_y.transform(y_test.reshape(-1,1)).flatten()

train_ds = TensorDataset(torch.tensor(X_train_scaled), torch.tensor(y_train_scaled).unsqueeze(1))
val_ds   = TensorDataset(torch.tensor(X_val_scaled), torch.tensor(y_val_scaled).unsqueeze(1))
test_ds  = TensorDataset(torch.tensor(X_test_scaled), torch.tensor(y_test_scaled).unsqueeze(1))

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_ds, batch_size=32)
test_loader  = DataLoader(test_ds, batch_size=32)

# ============================================================
# 8. Improved TFT-Like Model
# ============================================================
class GatedResidualNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, dropout=0.1):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.elu = nn.ELU()
        self.fc2 = nn.Linear(hidden_dim, input_dim)
        self.gate = nn.Linear(input_dim, input_dim)
        self.norm = nn.LayerNorm(input_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        residual = x
        x = self.elu(self.fc1(x))
        x = self.dropout(self.fc2(x))
        gate = torch.sigmoid(self.gate(x))
        x = gate * x + (1 - gate) * residual
        return self.norm(x)

class TFT_LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim=128, dropout=0.4):
        super().__init__()
        self.input_proj = nn.Linear(input_dim, hidden_dim)

        self.grn = GatedResidualNetwork(hidden_dim, hidden_dim, dropout)

        self.lstm1 = nn.LSTM(hidden_dim, hidden_dim, batch_first=True)
        self.lstm2 = nn.LSTM(hidden_dim, hidden_dim, batch_first=True)

        self.attn = nn.MultiheadAttention(hidden_dim, num_heads=4, batch_first=True)

        self.grn_out = GatedResidualNetwork(hidden_dim, hidden_dim, dropout)
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        x = self.input_proj(x)
        x = self.grn(x)

        x, _ = self.lstm1(x)
        x, _ = self.lstm2(x)

        attn_out, _ = self.attn(x, x, x)
        x = self.grn_out(x + attn_out)

        pooled = x.mean(dim=1)
        return self.fc(pooled)

model = TFT_LSTMModel(X_train.shape[2])
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# ============================================================
# 9. Training with Early Stopping
# ============================================================
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=2e-4)
epochs = 80
patience = 10
best_val = float("inf")
counter = 0

for epoch in range(epochs):

    model.train()
    train_losses = []
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())

    model.eval()
    val_losses = []
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            preds = model(xb)
            val_losses.append(criterion(preds, yb).item())

    avg_train = np.mean(train_losses)
    avg_val = np.mean(val_losses)

    print(f"Epoch {epoch+1} | Train: {avg_train:.4f} | Val: {avg_val:.4f}")

    if avg_val < best_val - 1e-4:
        best_val = avg_val
        counter = 0
        torch.save(model.state_dict(), "best_tft_lstm.pth")
    else:
        counter += 1
        if counter >= patience:
            print("Early stopping.")
            break

# ============================================================
# 10. Evaluation on Test Set
# ============================================================
model.load_state_dict(torch.load("best_tft_lstm.pth"))
model.eval()

y_true = []
y_pred = []

with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device)
        preds = model(xb)
        y_pred.append(preds.cpu().numpy())
        y_true.append(yb.numpy())

y_true = np.concatenate(y_true)
y_pred = np.concatenate(y_pred)

y_true = scaler_y.inverse_transform(y_true)
y_pred = scaler_y.inverse_transform(y_pred)

mse = np.mean((y_true - y_pred)**2)
rmse = np.sqrt(mse)
mae = np.mean(np.abs(y_true - y_pred))

mask = y_true.flatten() > 10
mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
smape = np.mean(2 * np.abs(y_pred - y_true) /
               (np.abs(y_true) + np.abs(y_pred) + 1e-8)) * 100

print("\n====================== RESULTS ======================")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"MAPE: {mape:.2f}%")
print(f"SMAPE: {smape:.2f}%")
print("=====================================================")
```
